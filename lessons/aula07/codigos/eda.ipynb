### 1.3.4 EDA Manually

```python
# There are duplicated rows
df.duplicated().sum()
```

```python
# Delete duplicated rows
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
# what the sex column can help us?
pd.crosstab(df.high_income,df.sex,margins=True,normalize=False)
```

```python
# income vs [sex & race]?
pd.crosstab(df.high_income,[df.sex,df.race],margins=True)
```

```python
%matplotlib inline

sns.catplot(x="sex", 
            hue="race", 
            col="high_income",
            data=df, kind="count",
            height=4, aspect=.7)
plt.show()
```

```python
g = sns.catplot(x="sex", 
                hue="workclass", 
                col="high_income",
                data=df, kind="count",
                height=4, aspect=.7)

g.savefig("HighIncome_Sex_Workclass.png", dpi=100)

run.log(
        {
            "High_Income vs Sex vs Workclass": wandb.Image("HighIncome_Sex_Workclass.png")
        }
    )
```
